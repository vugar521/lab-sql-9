**1. How many distinct (different) actors' last names are there?**

```python
query = """SELECT count(distinct last_name) AS dis_num
FROM actor;"""
last_name= pd.read_sql_query (query, engine)
last_name.head()
```

**2. Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.**

```python
query = '''SELECT *, 
CASE
WHEN weekday(rental_date)< 6 THEN 'workday'
ELSE 'weekend'
END AS 'day_type'
FROM rental;'''
day= pd.read_sql_query (query, engine)
day.head()
```

3. Get all films with ARMAGEDDON in the title.

```python
query = 'SELECT *\
FROM sakila.film \
WHERE title LIKE "%%ARMAGEDDON%%"'
ARMAGEDDON = pd.read_sql_query(query, engine)
ARMAGEDDON.head()
```

**4. Get 10 the longest films.**

```python
query = '''select f.title, sum(length) as max_length from film as f
join film_category as fc on f.film_id = fc.film_id
join category as c on fc.category_id=c.category_id
group by f.title
order by max_length desc
limit 10''';

longest_film = pd.read_sql_query(query, engine)
longest_film.head()
```

**5. How many films include Behind the Scenes content?**

```python
query = '''select * from film
WHERE special_features LIKE "%%Behind the Scenes%%";'''

behind = pd.read_sql_query(query, engine)
behind.head()
```

**6. Which kind of movies (rating) have a mean duration of more than two hours?**

```python
query = '''SELECT rating, AVG(length) AS mean_duration
FROM film
GROUP BY rating
HAVING AVG(length)>120;'''

mean_duration = pd.read_sql_query(query,engine)
mean_duration.head()
```

**7. Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.**

```python
query = '''SELECT title, length,
rank() OVER (ORDER BY length DESC) as 'rank'
FROM film
WHERE ISNULL(length) IS NOT NULL OR ISNULL(length) <> 0;'''

film_ranking = pd.read_sql_query(query,engine)
film_ranking.head(20)
```
